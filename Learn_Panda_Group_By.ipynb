# GROUP BY -  USING LOOPING

```python
# Using Group By, The complex method
```

```python

last200_rows=df.iloc[:-200,:]

# grouping 2 columns: by artist and year
# the result is an iterable object
grp=last200_rows.groupby(["artist"])
```

```python
#only print those artists with more than 4 work in one year
# you can then iterate through each grouping of artist x year

#note only grp_df is useful, grp_name is not that useful,we are know what we are grouping by already
for grp_name, grp_df in grp:
    df_local = grp_df.loc[:,["artist", "year", "short_title" ]]
    if len(df) > 4:
        print("{}\nrow count:{} \n---".format(df_local, len(df_local)))
    
```

```python
# Using AGGREGATE function on top of the GroupBy Object

# main idea: loop through the group and apply your aggregation process

min_year = pd.DataFrame(columns=["id", "artist", "earliest_year"], index=["id"])

for grp_name, grp_df in grp:
    
    earliest_year = grp_df["year"].min()
    earliest_year_id = grp_df["year"].idxmin()
    earliest_year = int(earliest_year) if earliest_year.is_integer() else 0
    min_year = min_year.append({"id": earliest_year_id, "artist": grp_name, "earliest_year": earliest_year}, ignore_index=True)
    #print("{} earliest year is {}".format(grp_name, earliest_year))

```

# GROUP BY & AGGREGATE FUNCTIONS

```python
# using AGGREGATE Functions the easy way with Group By

#we are using the numpy min function and apply it inside the groupby object

# Note: using the groupby.agg() function

easy_grp=last200_rows.groupby("artist").agg(np.min)
```

```python
# using AGGREGATE Functions the easy way with Group By: Take 2

# you can use panda's min function, which is an short-cut to numpy's min function

easy_grp=last200_rows.groupby("artist").min()
```

```python
# Now use FILTERING on Groupby Objects

# you need to declare the lambda function to do the filtering on the groupBy object

filter_title_less_than_10 = lambda grp_frm: len(grp_frm.name) < 7



last200_rows.groupby("title").filter(filter_title_less_than_10)
```

# GROUP BY AND  FILTERING

```python
# Now use FILTERING on Groupby Objects

# you need to declare the lambda function to do the filtering on the groupBy object

filter_title_less_than_10 = lambda gf: len(gf["title"].iloc[0]) ==10

# filter will apply the function to each groups of the GroupBy
# so here, each group as the same title, so just need to get the first row of the each group.
# To get the first row, I use iloc function to get [0], which means the first row, regardless of what is the index
# within the group[i]

my_grp = last200_rows.groupby("title")
my_grp.filter(filter_title_less_than_10)["title"].sort_values(ascending=False).unique().tolist()
```
